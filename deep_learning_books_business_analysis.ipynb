```python
# Step 1: Import required libraries

import pandas as pd
import numpy as np
import random

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)

print("Libraries imported successfully!")
```

```python
# Step 2: Create a list of deep learning books

books = [
    "Deep Learning - Goodfellow",
    "Deep Learning with Python - Chollet",
    "Hands-On Machine Learning - Geron",
    "Grokking Deep Learning - Trask",
    "Deep Learning from Scratch - Weidman",
    "Neural Networks and Deep Learning - Nielsen",
    "Fundamentals of Deep Learning - Buduma",
    "Generative Deep Learning - Foster",
    "Deep Learning with PyTorch - Stevens",
    "Dive into Deep Learning - Zhang"
]

authors = [
    "Ian Goodfellow",
    "Francois Chollet",
    "Aurelien Geron",
    "Andrew Trask",
    "Seth Weidman",
    "Michael Nielsen",
    "Nithin Buduma",
    "David Foster",
    "Eli Stevens",
    "Aston Zhang"
]

print("Books and authors created!")
```

```python
# Step 3: Define demographics

age_groups = ["18-25", "26-35", "36-45", "46-60", "60+"]
genders = ["Male", "Female", "Other"]

countries = [
    "UK", "USA", "Canada", "Germany", "India",
    "France", "Australia", "Netherlands"
]

customer_types = ["Student", "Professional", "Researcher", "Engineer"]
```

```python
# Step 4: Generate synthetic dataset

n_rows = 1000

data = {
    "Book": np.random.choice(books, n_rows),
    "Author": np.random.choice(authors, n_rows),
    "Units_Sold": np.random.randint(1, 50, n_rows),
    "Price_per_Unit": np.random.uniform(30, 80, n_rows).round(2),
    "Age_Group": np.random.choice(age_groups, n_rows),
    "Gender": np.random.choice(genders, n_rows),
    "Country": np.random.choice(countries, n_rows),
    "Customer_Type": np.random.choice(customer_types, n_rows)
}

df = pd.DataFrame(data)

df.head()
```

```python
# Step 5: Add financial columns

# Total Sales
df["Total_Sales"] = df["Units_Sold"] * df["Price_per_Unit"]

# Cost per unit (assume 40% to 70% of price)
df["Cost_per_Unit"] = df["Price_per_Unit"] * np.random.uniform(0.4, 0.7, n_rows)

# Total Cost
df["Total_Cost"] = df["Cost_per_Unit"] * df["Units_Sold"]

# Profit
df["Profit"] = df["Total_Sales"] - df["Total_Cost"]

# Profit Margin %
df["Profit_Margin (%)"] = (df["Profit"] / df["Total_Sales"]) * 100

df.head()
```

```python
# Step 6: Dataset overview

df.info()

df.describe()
```

```python
# Step 7: Check missing values

df.isnull().sum()
```

```python
# Step 8: Save dataset

df.to_csv("deep_learning_books_sales_dataset.csv", index=False)

print("Dataset saved successfully!")
```

```python
# Step 9: Total sales by book

sales_by_book = df.groupby("Book")["Total_Sales"].sum().sort_values(ascending=False)

sales_by_book
```

```python
# Step 9: Total sales by book

sales_by_book = df.groupby("Book")["Total_Sales"].sum().sort_values(ascending=False)

sales_by_book
```

```python
# Step 10: Profit by country

profit_by_country = df.groupby("Country")["Profit"].sum().sort_values(ascending=False)

profit_by_country
```

```python
# Step 11: Plot sales by book

plt.figure(figsize=(12,6))

sales_by_book.plot(kind="bar")

plt.title("Total Sales by Deep Learning Book")
plt.xlabel("Book")
plt.ylabel("Sales")
plt.xticks(rotation=45)

plt.show()
```

```python
# Step 12: Profit margin distribution

plt.figure(figsize=(10,6))

sns.histplot(df["Profit_Margin (%)"], bins=30)

plt.title("Profit Margin Distribution")
plt.xlabel("Profit Margin (%)")

plt.show()
```

```python
# Step 13: Sales by age group

sales_by_age = df.groupby("Age_Group")["Total_Sales"].sum()

sales_by_age.plot(kind="pie", autopct="%1.1f%%")

plt.title("Sales Distribution by Age Group")

plt.show()
```
