```python
# Step 1: Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# For reproducibility
np.random.seed(42)

print("Libraries imported successfully!")
```

```python
# Step 2: Define Crash Course Book Information

books = [
    "Crash Course Data Science",
    "Crash Course Machine Learning",
    "Crash Course Python for Data Analysis",
    "Crash Course Statistics for ML",
    "Crash Course Deep Learning",
    "Crash Course AI Fundamentals",
    "Crash Course Data Visualization",
    "Crash Course Big Data Analytics"
]

price = [29.99, 34.99, 24.99, 27.99, 39.99, 32.99, 26.99, 36.99]

cost = [12.00, 15.00, 10.00, 11.50, 18.00, 14.50, 11.00, 17.00]
```

```python
# Step 3: Create Demographics Data

regions = ["UK", "USA", "Canada", "Germany", "India", "Australia"]

age_groups = ["18-24", "25-34", "35-44", "45-54"]

genders = ["Male", "Female"]
```

```python
# Step 4: Generate Synthetic Sales Data

data = []

for i in range(500):
    
    book_index = np.random.randint(0, len(books))
    
    book = books[book_index]
    
    selling_price = price[book_index]
    
    unit_cost = cost[book_index]
    
    quantity = np.random.randint(1, 10)
    
    revenue = selling_price * quantity
    
    total_cost = unit_cost * quantity
    
    profit = revenue - total_cost
    
    profit_margin = (profit / revenue) * 100
    
    region = np.random.choice(regions)
    
    age_group = np.random.choice(age_groups)
    
    gender = np.random.choice(genders)
    
    data.append([
        book,
        selling_price,
        unit_cost,
        quantity,
        revenue,
        total_cost,
        profit,
        profit_margin,
        region,
        age_group,
        gender
    ])
```

```python
# Step 5: Create DataFrame

columns = [
    "Book",
    "Selling Price",
    "Unit Cost",
    "Quantity Sold",
    "Revenue",
    "Total Cost",
    "Profit",
    "Profit Margin (%)",
    "Region",
    "Age Group",
    "Gender"
]

df = pd.DataFrame(data, columns=columns)

df.head()
```

```python
# Step 5: Create DataFrame

columns = [
    "Book",
    "Selling Price",
    "Unit Cost",
    "Quantity Sold",
    "Revenue",
    "Total Cost",
    "Profit",
    "Profit Margin (%)",
    "Region",
    "Age Group",
    "Gender"
]

df = pd.DataFrame(data, columns=columns)

df.head()
```

```python
# Step 6: Save dataset

df.to_csv("crash_course_books_sales.csv", index=False)

print("Dataset saved successfully!")
```

```python
# Step 7: Basic analysis

print("Total Revenue:", df["Revenue"].sum())

print("Total Profit:", df["Profit"].sum())

print("Average Profit Margin:", df["Profit Margin (%)"].mean())
```

```python
# Step 8: Sales by Region

region_sales = df.groupby("Region")["Revenue"].sum()

print(region_sales)
```

```python
# Step 9: Profit margin by book

book_profit = df.groupby("Book")["Profit Margin (%)"].mean()

print(book_profit)
```

```python
# Step 10: Plot revenue by book

book_revenue = df.groupby("Book")["Revenue"].sum()

book_revenue.plot(kind="bar")

plt.title("Revenue by Crash Course Book")

plt.ylabel("Revenue")

plt.xlabel("Book")

plt.show()
```
