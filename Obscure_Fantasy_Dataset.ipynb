```python
# Step 2: Import libraries

import pandas as pd
import numpy as np
import random

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)
```

```python
# Step 3: Obscure fantasy book titles and authors

book_titles = [
    "The Forgotten Rune",
    "Shadow of the Crystal Keep",
    "Whispers of the Iron Forest",
    "The Last Dragon Scribe",
    "Moonblade Prophecy",
    "Empire of Hollow Gods",
    "The Cursed Ember Crown",
    "Oracle of the Broken Star",
    "The Silent Mage of Eldoria",
    "Chronicles of the Ashen Throne",
    "The Phantom Spellbinder",
    "The Obsidian Gatekeeper",
    "Sword of Eternal Twilight",
    "The Necromancer's Garden",
    "The Frozen Realm Awakens",
    "The Wandering Alchemist",
    "Echoes of the Forgotten Realm",
    "The Void King's Apprentice",
    "Realm of Crimson Magic",
    "The Lost Kingdom of Nyx"
]

authors = [
    "Aelric Moonshadow",
    "Lyra Nightwind",
    "Thorin Blackthorn",
    "Elowen Frost",
    "Darius Grimveil",
    "Seraphina Dusk",
    "Kael Stormborn",
    "Isolde Ravencrest",
    "Lucian Darkwater",
    "Rowan Ashbinder"
]
```

```python
# Step 4: Demographic categories

genders = ["Male", "Female", "Non-binary"]
age_groups = ["18-24", "25-34", "35-44", "45-54", "55+"]
regions = ["UK", "USA", "Canada", "Australia", "Germany", "France"]
income_levels = ["Low", "Medium", "High"]
formats = ["Paperback", "Hardcover", "Ebook"]
```

```python
# Step 5: Generate dataset

num_records = 500

data = []

for i in range(num_records):

    title = random.choice(book_titles)
    author = random.choice(authors)

    price = round(np.random.uniform(5, 30), 2)
    cost = round(price * np.random.uniform(0.3, 0.7), 2)

    units_sold = np.random.randint(100, 5000)

    revenue = round(price * units_sold, 2)
    total_cost = round(cost * units_sold, 2)

    profit = round(revenue - total_cost, 2)
    profit_margin = round((profit / revenue) * 100, 2)

    marketing_spend = round(np.random.uniform(500, 10000), 2)

    record = {
        "Book_ID": i+1,
        "Title": title,
        "Author": author,
        "Price": price,
        "Cost": cost,
        "Units_Sold": units_sold,
        "Revenue": revenue,
        "Total_Cost": total_cost,
        "Profit": profit,
        "Profit_Margin_%": profit_margin,
        "Marketing_Spend": marketing_spend,
        "Gender": random.choice(genders),
        "Age_Group": random.choice(age_groups),
        "Region": random.choice(regions),
        "Income_Level": random.choice(income_levels),
        "Format": random.choice(formats)
    }

    data.append(record)
```

```python
# Step 6: Convert to DataFrame

df = pd.DataFrame(data)

# Display first 5 rows
df.head()
```

```python
# Step 7: Dataset info

df.info()

df.describe()
```

```python
# Step 8: Save to CSV

df.to_csv("obscure_fantasy_books_dataset.csv", index=False)

print("Dataset saved successfully.")
```
