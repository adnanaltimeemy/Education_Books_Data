```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

print("Libraries imported successfully")
```

```python
# Step 2: Define number of records
num_records = 5000

print(f"Dataset will contain {num_records} records")
```

```python
# Step 3: O'Reilly book categories
book_titles = [
    "Python for Data Science",
    "Machine Learning with Python",
    "Deep Learning Essentials",
    "Data Engineering Fundamentals",
    "SQL for Business Analytics",
    "AI for Business Leaders",
    "Statistics for Data Scientists",
    "Cloud Computing with AWS",
    "Big Data Analytics",
    "Natural Language Processing"
]

categories = [
    "Data Science",
    "Machine Learning",
    "AI",
    "Business",
    "Cloud",
    "Programming",
    "Analytics"
]

regions = ["North America", "Europe", "Asia", "UK", "Australia"]
genders = ["Male", "Female", "Other"]
education_levels = ["Bachelor", "Master", "PhD", "High School"]
```

```python
# Step 4: Create demographics data

customer_id = np.arange(10000, 10000 + num_records)

age = np.random.randint(18, 65, num_records)

gender = np.random.choice(genders, num_records)

education = np.random.choice(education_levels, num_records)

region = np.random.choice(regions, num_records)
```

```python
# Step 5: Generate sales data

book = np.random.choice(book_titles, num_records)

category = np.random.choice(categories, num_records)

units_sold = np.random.randint(1, 20, num_records)

price = np.round(np.random.uniform(20, 120, num_records), 2)

cost = np.round(price * np.random.uniform(0.4, 0.7, num_records), 2)

revenue = price * units_sold

total_cost = cost * units_sold
```

```python
# Step 6: Calculate profit metrics

profit = revenue - total_cost

profit_margin = np.round((profit / revenue) * 100, 2)
```

```python
# Step 7: Create sales dates

dates = pd.date_range(start="2022-01-01", end="2025-12-31")

sale_date = np.random.choice(dates, num_records)
```

```python
# Step 8: Combine all into DataFrame

df = pd.DataFrame({

    "Customer_ID": customer_id,
    "Age": age,
    "Gender": gender,
    "Education": education,
    "Region": region,

    "Book_Title": book,
    "Category": category,

    "Units_Sold": units_sold,
    "Price": price,
    "Cost": cost,

    "Revenue": revenue,
    "Total_Cost": total_cost,

    "Profit": profit,
    "Profit_Margin_%": profit_margin,

    "Sale_Date": sale_date
})

df.head()
```

```python
# Step 9: Explore dataset

print("Dataset shape:")
print(df.shape)

print("\nDataset info:")
print(df.info())

print("\nSummary statistics:")
print(df.describe())
```

```python
# Step 10: Save dataset

df.to_csv("oreilly_books_sales_dataset.csv", index=False)

print("Dataset saved successfully")
```

```python
total_revenue = df["Revenue"].sum()
print(f"Total Revenue: ${total_revenue:,.2f}")
```
