```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# Make plots display in notebook
%matplotlib inline
```

```python
# Step 2: Define columns for AI book sales dataset

columns = [
    "Order_ID",
    "Book_Title",
    "Author",
    "Genre",
    "Customer_Age",
    "Customer_Gender",
    "Country",
    "Purchase_Date",
    "Units_Sold",
    "Sale_Price",
    "Cost_Price",
    "Revenue",
    "Cost",
    "Profit",
    "Profit_Margin"
]

print("Columns created successfully")
```

```python
# Step 3: Create sample data lists

book_titles = [
    "AI Basics",
    "Machine Learning Guide",
    "Deep Learning Mastery",
    "Python for AI",
    "AI in Business",
    "Neural Networks Explained",
    "AI Ethics Handbook",
    "Computer Vision Essentials",
    "Natural Language Processing",
    "AI Automation"
]

authors = [
    "John Smith",
    "Sarah Johnson",
    "David Lee",
    "Emma Brown",
    "Michael Chen",
    "Sophia Davis"
]

genres = [
    "Beginner",
    "Advanced",
    "Business",
    "Technical"
]

countries = [
    "UK",
    "USA",
    "Canada",
    "Germany",
    "India",
    "Australia"
]

genders = [
    "Male",
    "Female"
]
```

```python
# Step 4: Generate 500 rows of data

data = []

for i in range(500):
    
    order_id = 1000 + i
    
    book = random.choice(book_titles)
    author = random.choice(authors)
    genre = random.choice(genres)
    
    age = random.randint(18, 70)
    gender = random.choice(genders)
    country = random.choice(countries)
    
    date = pd.Timestamp("2024-01-01") + pd.to_timedelta(random.randint(0, 365), unit="d")
    
    units = random.randint(1, 10)
    
    sale_price = round(random.uniform(15, 60), 2)
    cost_price = round(sale_price * random.uniform(0.4, 0.7), 2)
    
    revenue = round(units * sale_price, 2)
    cost = round(units * cost_price, 2)
    
    profit = round(revenue - cost, 2)
    
    profit_margin = round((profit / revenue) * 100, 2)
    
    row = [
        order_id,
        book,
        author,
        genre,
        age,
        gender,
        country,
        date,
        units,
        sale_price,
        cost_price,
        revenue,
        cost,
        profit,
        profit_margin
    ]
    
    data.append(row)

print("Dataset generated successfully")
```

```python
# Step 5: Convert to pandas DataFrame

df = pd.DataFrame(data, columns=columns)

df.head()
```

```python
# Step 6: Save dataset

df.to_csv("AI_BOOKS_SALES_DATASET.csv", index=False)

print("Dataset saved successfully")
```
