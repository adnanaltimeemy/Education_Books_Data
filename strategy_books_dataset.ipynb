```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside notebook
%matplotlib inline
```

```python
# Step 2: Create sample dataset

np.random.seed(42)

n = 500  # number of records

data = {
    "Book_ID": range(1000, 1000+n),
    
    "Book_Title": np.random.choice([
        "Business Strategy 101",
        "Corporate Leadership",
        "Competitive Advantage",
        "Market Domination",
        "Strategic Thinking",
        "Growth Strategy Guide"
    ], n),
    
    "Author": np.random.choice([
        "Michael Porter",
        "John Maxwell",
        "Peter Drucker",
        "Simon Sinek",
        "Clayton Christensen"
    ], n),
    
    "Customer_Age": np.random.randint(18, 65, n),
    
    "Customer_Gender": np.random.choice(["Male", "Female"], n),
    
    "Customer_Income": np.random.choice(
        ["Low", "Medium", "High"], n
    ),
    
    "Units_Sold": np.random.randint(1, 50, n),
    
    "Unit_Price": np.round(np.random.uniform(10, 50, n), 2),
    
    "Cost_Per_Unit": np.round(np.random.uniform(5, 30, n), 2),
    
    "Region": np.random.choice([
        "North", "South", "East", "West"
    ], n)
}

df = pd.DataFrame(data)

df.head()
```

```python
# Step 3: Calculate financial metrics

df["Total_Sales"] = df["Units_Sold"] * df["Unit_Price"]

df["Total_Cost"] = df["Units_Sold"] * df["Cost_Per_Unit"]

df["Profit"] = df["Total_Sales"] - df["Total_Cost"]

df["Profit_Margin"] = (df["Profit"] / df["Total_Sales"]) * 100

df.head()
```

```python
# Step 4: Dataset info

print("Dataset Shape:", df.shape)

df.info()

df.describe()
```

```python
# Step 5: Missing values

df.isnull().sum()
```

```python
# Step 6: Overall performance

total_sales = df["Total_Sales"].sum()
total_profit = df["Profit"].sum()

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
# Step 7: Sales grouped by Book

sales_by_book = df.groupby("Book_Title")["Total_Sales"].sum()

sales_by_book.sort_values(ascending=False)
```

```python
# Step 8: Profit margin by Book

profit_margin_book = df.groupby("Book_Title")["Profit_Margin"].mean()

profit_margin_book.sort_values(ascending=False)
```

```python
# Step 9: Gender analysis

gender_sales = df.groupby("Customer_Gender")["Total_Sales"].sum()

gender_sales
```

```python
# Step 10: Income analysis

income_sales = df.groupby("Customer_Income")["Total_Sales"].sum()

income_sales
```

```python
# Step 11: Visualization

plt.figure(figsize=(10,6))

sales_by_book.plot(kind="bar")

plt.title("Total Sales by Strategy Book")
plt.ylabel("Sales")
plt.show()
```

```python
# Step 12: Profit Margin Distribution

plt.figure(figsize=(10,6))

sns.histplot(df["Profit_Margin"], bins=20)

plt.title("Profit Margin Distribution")
plt.show()
```

```python
# Step 13: Age vs Sales

plt.figure(figsize=(10,6))

sns.scatterplot(x=df["Customer_Age"], y=df["Total_Sales"])

plt.title("Customer Age vs Sales")
plt.show()
```

```python
# Step 14: Region analysis

region_sales = df.groupby("Region")["Total_Sales"].sum()

region_sales.plot(kind="bar", figsize=(8,5))

plt.title("Sales by Region")
plt.show()
```

```python
# Step 15: Save dataset to CSV

df.to_csv("strategy_books_dataset.csv", index=False)

print("Dataset saved successfully.")
```
