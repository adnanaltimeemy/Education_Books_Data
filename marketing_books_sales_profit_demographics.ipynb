```python
# Step 1: Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Step 2: Define number of records

num_records = 1000
```

```python
# Step 3: Create book-related variables

book_ids = np.arange(1, num_records+1)

book_titles = [
    "Digital Marketing 101", "SEO Mastery", "Social Media Strategy",
    "Marketing Analytics", "Brand Management", "Email Marketing Guide",
    "Content Marketing", "Advertising Strategy", "Consumer Psychology",
    "Marketing Research"
]

book_title = np.random.choice(book_titles, num_records)

categories = ["Digital", "SEO", "Social Media", "Analytics", "Branding"]
category = np.random.choice(categories, num_records)

publication_year = np.random.randint(2015, 2026, num_records)
```

```python
# Step 4: Customer demographic variables

age_group = np.random.choice(
    ["18-25", "26-35", "36-45", "46-55", "56+"], num_records
)

gender = np.random.choice(["Male", "Female"], num_records)

income_level = np.random.choice(
    ["Low", "Medium", "High"], num_records
)

region = np.random.choice(
    ["London", "Manchester", "Birmingham", "Leeds", "Glasgow"],
    num_records
)
```

```python
# Step 5: Marketing variables

marketing_channel = np.random.choice(
    ["Facebook", "Google Ads", "Email", "Instagram", "Organic"],
    num_records
)

ad_spend = np.round(np.random.uniform(5, 50, num_records), 2)
```

```python
# Step 6: Financial variables

units_sold = np.random.randint(10, 500, num_records)

price = np.round(np.random.uniform(10, 60, num_records), 2)

cost = np.round(price * np.random.uniform(0.4, 0.8, num_records), 2)

sales_revenue = units_sold * price

total_cost = units_sold * cost
```

```python
# Step 7: Profit calculations

profit = sales_revenue - total_cost

profit_margin = (profit / sales_revenue) * 100
```

```python
# Step 8: Create dataset

df = pd.DataFrame({
    
    "Book_ID": book_ids,
    "Book_Title": book_title,
    "Category": category,
    "Publication_Year": publication_year,
    
    # Demographics
    "Age_Group": age_group,
    "Gender": gender,
    "Income_Level": income_level,
    "Region": region,
    
    # Marketing
    "Marketing_Channel": marketing_channel,
    "Ad_Spend": ad_spend,
    
    # Sales
    "Units_Sold": units_sold,
    "Price": price,
    "Cost": cost,
    
    # Financial metrics
    "Revenue": sales_revenue,
    "Total_Cost": total_cost,
    "Profit": profit,
    "Profit_Margin (%)": profit_margin
})

df.head()
```

```python
# Step 9: Save dataset

df.to_csv("marketing_books_dataset.csv", index=False)

print("Dataset saved successfully!")
```

```python
# Step 10: Dataset info

df.info()

df.describe()
```

```python
# Step 11: Sales by marketing channel

channel_sales = df.groupby("Marketing_Channel")["Revenue"].sum()

print(channel_sales)
```

```python
# Step 12: Visualization

plt.figure(figsize=(10,6))

sns.barplot(
    x="Region",
    y="Profit",
    data=df
)

plt.title("Profit by Region")
plt.show()
```

```python
# Step 13: Profit margin histogram

plt.figure(figsize=(10,6))

sns.histplot(df["Profit_Margin (%)"], bins=30)

plt.title("Profit Margin Distribution")
plt.show()
```
