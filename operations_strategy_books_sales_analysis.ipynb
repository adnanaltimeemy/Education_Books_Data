```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import random

# Set seed for reproducibility
random.seed(42)
np.random.seed(42)

print("Libraries imported successfully!")
```

```python
# Step 2: Define operations strategy books

books = [
    "Operations Strategy – Nigel Slack",
    "The Goal – Eliyahu Goldratt",
    "Lean Thinking – Womack & Jones",
    "The Toyota Way – Jeffrey Liker",
    "Operations Management – Stevenson",
    "The Phoenix Project – Gene Kim",
    "Operations Strategy – Michael Lewis",
    "Supply Chain Management – Chopra",
    "Strategic Operations Management – Brown",
    "Operations Management: Theory and Practice – Mahadevan"
]

print("Books list created!")
```

```python
# Step 3: Define demographic categories

age_groups = ["18-24", "25-34", "35-44", "45-54", "55+"]
genders = ["Male", "Female", "Non-binary"]
regions = ["UK", "USA", "Canada", "Europe", "Asia", "Australia"]

education_levels = [
    "Undergraduate Student",
    "MBA Student",
    "PhD Student",
    "Professor",
    "Industry Professional"
]

print("Demographics defined!")
```

```python
# Step 4: Generate dataset

num_records = 500

data = []

for i in range(num_records):

    book = random.choice(books)

    units_sold = random.randint(10, 500)

    price = round(random.uniform(20, 120), 2)

    revenue = units_sold * price

    cost = revenue * random.uniform(0.4, 0.7)

    profit = revenue - cost

    profit_margin = (profit / revenue) * 100

    record = {
        "Book": book,
        "Units_Sold": units_sold,
        "Price": price,
        "Revenue": round(revenue, 2),
        "Cost": round(cost, 2),
        "Profit": round(profit, 2),
        "Profit_Margin (%)": round(profit_margin, 2),
        "Age_Group": random.choice(age_groups),
        "Gender": random.choice(genders),
        "Region": random.choice(regions),
        "Education_Level": random.choice(education_levels)
    }

    data.append(record)

print("Dataset generated!")
```

```python
# Step 5: Create DataFrame

df = pd.DataFrame(data)

df.head()
```

```python
# Step 6: Dataset overview

print("Shape of dataset:", df.shape)

df.info()

df.describe()
```

```python
# Step 7: Save dataset

df.to_csv("operations_strategy_books_dataset.csv", index=False)

print("Dataset saved successfully!")
```

```python
# Step 8: Sales by book

sales_by_book = df.groupby("Book")["Revenue"].sum().sort_values(ascending=False)

sales_by_book
```

```python
# Step 9: Profit margin by book

profit_margin = df.groupby("Book")["Profit_Margin (%)"].mean().sort_values(ascending=False)

profit_margin
```

```python
# Step 10: Sales by age group

sales_by_age = df.groupby("Age_Group")["Revenue"].sum()

sales_by_age
```

```python
# Step 11: Sales by region

sales_by_region = df.groupby("Region")["Revenue"].sum()

sales_by_region
```

```python
# Step 12: Sales by education level

sales_by_education = df.groupby("Education_Level")["Revenue"].sum()

sales_by_education
```
