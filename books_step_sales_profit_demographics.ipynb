```python
# Step 1: Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

print("Libraries imported successfully.")
```

```python
# Step 2: Define dataset columns

columns = [
    "Transaction_ID",
    "Date",
    "Book_Title",
    "Category",
    "Units_Sold",
    "Price_Per_Unit",
    "Cost_Per_Unit",
    "Revenue",
    "Cost",
    "Profit",
    "Profit_Margin",
    
    # Demographics
    "Customer_ID",
    "Customer_Age",
    "Customer_Gender",
    "Customer_Income_Level",
    "Customer_Region",
    "Customer_Education"
]

print("Columns defined successfully.")
```

```python
# Step 3: Generate synthetic dataset

np.random.seed(42)

n = 500

data = {
    "Transaction_ID": range(1, n+1),
    
    "Date": pd.date_range(start="2025-01-01", periods=n, freq="D"),
    
    "Book_Title": np.random.choice([
        "Community Leadership",
        "Social Impact Guide",
        "Education for Change",
        "Sustainable Future",
        "Youth Empowerment",
        "Women in Leadership",
        "Ethical Business",
        "Climate Action"
    ], n),
    
    "Category": np.random.choice([
        "Education",
        "Leadership",
        "Environment",
        "Social Justice"
    ], n),
    
    "Units_Sold": np.random.randint(1, 20, n),
    
    "Price_Per_Unit": np.random.uniform(8, 30, n).round(2),
    
    "Cost_Per_Unit": np.random.uniform(4, 15, n).round(2),
    
    "Customer_ID": np.random.randint(1000, 2000, n),
    
    "Customer_Age": np.random.randint(18, 70, n),
    
    "Customer_Gender": np.random.choice([
        "Male", "Female", "Non-binary"
    ], n),
    
    "Customer_Income_Level": np.random.choice([
        "Low", "Medium", "High"
    ], n),
    
    "Customer_Region": np.random.choice([
        "London", "Manchester", "Birmingham", "Leeds", "Glasgow"
    ], n),
    
    "Customer_Education": np.random.choice([
        "High School", "Bachelor", "Master", "PhD"
    ], n)
}

df = pd.DataFrame(data)

df.head()
```

```python
# Step 4: Calculate financial metrics

df["Revenue"] = df["Units_Sold"] * df["Price_Per_Unit"]

df["Cost"] = df["Units_Sold"] * df["Cost_Per_Unit"]

df["Profit"] = df["Revenue"] - df["Cost"]

df["Profit_Margin"] = (df["Profit"] / df["Revenue"]).round(2)

df.head()
```

```python
# Step 5: Save dataset

df.to_csv("books_step_social_enterprise_dataset.csv", index=False)

print("Dataset saved successfully.")
```
