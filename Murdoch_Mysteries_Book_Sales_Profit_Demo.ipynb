```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from faker import Faker

fake = Faker()

# Set random seed for reproducibility
np.random.seed(42)
```

```python
books = [
    ("Except the Dying", 1997),
    ("Under the Dragon's Tail", 1998),
    ("Poor Tom is Cold", 2001),
    ("Let Loose the Dogs", 2003),
    ("Night's Child", 2005),
    ("Vices of My Blood", 2006),
    ("A Journeyman to Grief", 2007),
    ("Shipwreck", 2011),
    ("Let Darkness Bury the Dead", 2017)
]

books_df = pd.DataFrame(books, columns=["Book_Title", "Year"])
books_df
```

```python
books_df["Units_Sold"] = np.random.randint(50000, 500000, size=len(books_df))

books_df["Price_per_Unit"] = np.random.uniform(8.99, 19.99, size=len(books_df)).round(2)

books_df["Revenue"] = books_df["Units_Sold"] * books_df["Price_per_Unit"]

books_df
```

```python
books_df["Cost_per_Unit"] = np.random.uniform(3.00, 8.00, size=len(books_df)).round(2)

books_df["Total_Cost"] = books_df["Units_Sold"] * books_df["Cost_per_Unit"]

books_df["Profit"] = books_df["Revenue"] - books_df["Total_Cost"]

books_df["Profit_Margin"] = (books_df["Profit"] / books_df["Revenue"]).round(2)

books_df
```

```python
demographics = []

age_groups = ["18-25", "26-35", "36-45", "46-60", "60+"]
genders = ["Male", "Female", "Other"]
countries = ["Canada", "UK", "USA", "Australia", "Germany"]

for i in range(1000):
    demographics.append({
        "Reader_ID": i,
        "Age_Group": np.random.choice(age_groups),
        "Gender": np.random.choice(genders, p=[0.45, 0.50, 0.05]),
        "Country": np.random.choice(countries),
        "Book_Purchased": np.random.choice(books_df["Book_Title"])
    })

demo_df = pd.DataFrame(demographics)

demo_df.head()
```

```python
books_df.to_csv("murdoch_books_sales.csv", index=False)
demo_df.to_csv("murdoch_books_demographics.csv", index=False)

print("Datasets saved successfully.")
```
