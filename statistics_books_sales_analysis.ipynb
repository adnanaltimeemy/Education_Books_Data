```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries imported successfully!")
```

```python
# Step 2: Create sample dataset

np.random.seed(42)

n = 200   # number of records

data = {
    "Book_ID": np.arange(1001, 1001+n),
    
    "Book_Title": np.random.choice([
        "Statistics Fundamentals",
        "Advanced Statistics",
        "Probability Theory",
        "Statistical Methods",
        "Applied Statistics",
        "Statistics for Beginners",
        "Business Statistics",
        "Statistical Analysis Guide"
    ], n),
    
    "Author": np.random.choice([
        "Smith", "Johnson", "Williams", "Brown", "Jones", "Miller"
    ], n),
    
    "Customer_Age": np.random.randint(18, 65, n),
    
    "Customer_Gender": np.random.choice([
        "Male", "Female"
    ], n),
    
    "Customer_Income": np.random.randint(20000, 100000, n),
    
    "Book_Price": np.random.randint(20, 100, n),
    
    "Book_Cost": np.random.randint(10, 60, n),
    
    "Quantity_Sold": np.random.randint(1, 10, n),
    
    "Region": np.random.choice([
        "London", "Manchester", "Birmingham", "Leeds", "Glasgow"
    ], n)
}

df = pd.DataFrame(data)

df.head()
```

```python
# Step 3: Calculate financial metrics

df["Revenue"] = df["Book_Price"] * df["Quantity_Sold"]

df["Profit"] = (df["Book_Price"] - df["Book_Cost"]) * df["Quantity_Sold"]

df["Profit_Margin (%)"] = (df["Profit"] / df["Revenue"]) * 100

df.head()
```

```python
# Step 4: Save dataset

df.to_csv("statistics_books_sales_dataset.csv", index=False)

print("Dataset saved successfully!")
```

```python
# Step 5: Dataset overview

print("Dataset Shape:", df.shape)

print("\nDataset Info:")
print(df.info())

print("\nDataset Description:")
df.describe()
```

```python
# Step 6: Missing values

df.isnull().sum()
```

```python
# Step 7: Total revenue and profit

total_revenue = df["Revenue"].sum()
total_profit = df["Profit"].sum()

print("Total Revenue:", total_revenue)
print("Total Profit:", total_profit)
```

```python
# Step 8: Profit by book

profit_by_book = df.groupby("Book_Title")["Profit"].sum().sort_values(ascending=False)

profit_by_book
```

```python
# Step 9: Sales by gender

sales_by_gender = df.groupby("Customer_Gender")["Revenue"].sum()

sales_by_gender.plot(kind="bar", title="Sales by Gender")
plt.show()
```

```python
# Step 10: Profit margin distribution

sns.histplot(df["Profit_Margin (%)"], bins=20)

plt.title("Profit Margin Distribution")
plt.show()
```

```python
# Step 11: Sales vs Age

plt.scatter(df["Customer_Age"], df["Revenue"])

plt.title("Sales vs Customer Age")
plt.xlabel("Age")
plt.ylabel("Revenue")

plt.show()
```

```python
# Step 12: Sales by region

sales_region = df.groupby("Region")["Revenue"].sum()

sales_region.plot(kind="bar", title="Sales by Region")

plt.show()
```

```python
# Step 13: Correlation heatmap

numeric_df = df.select_dtypes(include=np.number)

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```
