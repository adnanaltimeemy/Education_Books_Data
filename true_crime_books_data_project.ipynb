```python
# Step 1: Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots appear in notebook
%matplotlib inline

# Set random seed for reproducibility
np.random.seed(42)

print("Libraries imported successfully.")
```

```python
# Step 2: Define possible values

titles = [
    "Silent Witness Files",
    "Serial Killer Mind",
    "Dark Case Chronicles",
    "Murder Investigation Files",
    "The Hidden Truth",
    "Cold Case Secrets",
    "Crime Scene Diaries",
    "Deadly Evidence",
    "Unsolved Crimes",
    "Justice Denied"
]

authors = [
    "John Carter",
    "Emily Stone",
    "Michael Reed",
    "Sarah Walker",
    "David Hunter"
]

regions = ["North America", "Europe", "Asia", "Australia"]

age_groups = ["18-24", "25-34", "35-44", "45-54", "55+"]

genders = ["Male", "Female"]

education_levels = [
    "High School",
    "Bachelor",
    "Master",
    "PhD"
]

print("Dataset categories defined.")
```

```python
# Step 3: Create synthetic dataset

n = 500  # number of records

data = {
    "Book_ID": np.arange(1, n+1),
    
    "Title": np.random.choice(titles, n),
    
    "Author": np.random.choice(authors, n),
    
    "Year": np.random.randint(2010, 2025, n),
    
    "Region": np.random.choice(regions, n),
    
    "Units_Sold": np.random.randint(500, 50000, n),
    
    "Price": np.round(np.random.uniform(10, 35, n), 2),
    
    "Reader_Age_Group": np.random.choice(age_groups, n),
    
    "Reader_Gender": np.random.choice(genders, n, p=[0.4, 0.6]),  # more female readers
    
    "Education_Level": np.random.choice(education_levels, n)
}

df = pd.DataFrame(data)

df.head()
```

```python
# Step 4: Financial calculations

# Revenue
df["Revenue"] = df["Units_Sold"] * df["Price"]

# Cost (assume cost is 40% to 70% of revenue)
df["Cost"] = df["Revenue"] * np.random.uniform(0.4, 0.7, len(df))

# Profit
df["Profit"] = df["Revenue"] - df["Cost"]

# Profit margin %
df["Profit_Margin"] = (df["Profit"] / df["Revenue"]) * 100

df.head()
```

```python
# Step 5: Dataset summary

print("Dataset shape:", df.shape)

df.info()

df.describe()
```

```python
# Step 6: Save dataset

df.to_csv("true_crime_books_dataset.csv", index=False)

print("Dataset saved successfully.")
```

```python
# Step 7: Total revenue by region

region_sales = df.groupby("Region")["Revenue"].sum()

region_sales.plot(kind="bar")

plt.title("Revenue by Region")
plt.ylabel("Revenue")
plt.xlabel("Region")

plt.show()
```

```python
# Step 8: Profit margin distribution

plt.hist(df["Profit_Margin"], bins=20)

plt.title("Profit Margin Distribution")
plt.xlabel("Profit Margin %")
plt.ylabel("Frequency")

plt.show()
```

```python
# Step 9: Gender analysis

gender_counts = df["Reader_Gender"].value_counts()

gender_counts.plot(kind="pie", autopct="%1.1f%%")

plt.title("Reader Gender Distribution")

plt.show()
```

```python
# Step 10: Age group revenue

age_sales = df.groupby("Reader_Age_Group")["Revenue"].sum()

age_sales.plot(kind="bar")

plt.title("Revenue by Age Group")

plt.show()
```

```python
# Step 11: Most profitable books

top_books = df.sort_values("Profit", ascending=False).head(10)

top_books[["Title", "Author", "Profit"]]
```

```python
# Step 12: Correlation heatmap

numeric_df = df[["Units_Sold", "Price", "Revenue", "Cost", "Profit", "Profit_Margin"]]

sns.heatmap(numeric_df.corr(), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```
