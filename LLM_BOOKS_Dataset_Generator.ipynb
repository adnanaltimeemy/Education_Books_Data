```python
# Step 1: Import required libraries

import pandas as pd
import numpy as np
import random

# For visualization later
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)
random.seed(42)
```

```python
# Step 2: Define number of records

num_records = 1000

print("Dataset size:", num_records)
```

```python
# Step 3: Create book data

book_titles = [
    "Intro to LLMs",
    "Advanced AI Systems",
    "Deep Learning with Python",
    "Transformers Explained",
    "AI Ethics and Society",
    "Natural Language Processing Guide",
    "Machine Learning Foundations",
    "LLM Fine-Tuning Handbook",
    "Generative AI Applications",
    "Prompt Engineering Mastery"
]

authors = [
    "John Smith",
    "Emily Johnson",
    "Michael Brown",
    "Sarah Davis",
    "David Wilson",
    "Sophia Taylor"
]

genres = [
    "Technology",
    "AI",
    "Programming",
    "Education",
    "Data Science"
]
```

```python
# Step 4: Customer demographic fields

genders = ["Male", "Female", "Other"]

regions = [
    "UK",
    "USA",
    "Canada",
    "Germany",
    "India",
    "Australia"
]

education_levels = [
    "High School",
    "Bachelor",
    "Master",
    "PhD"
]
```

```python
# Step 5: Generate numerical data

prices = np.round(np.random.uniform(10, 100, num_records), 2)

quantity_sold = np.random.randint(1, 20, num_records)

cost = prices * np.random.uniform(0.4, 0.8, num_records)

revenue = prices * quantity_sold

profit = revenue - (cost * quantity_sold)

profit_margin = (profit / revenue) * 100
```

```python
# Step 6: Generate demographic variables

customer_age = np.random.randint(18, 70, num_records)

customer_gender = np.random.choice(genders, num_records)

customer_region = np.random.choice(regions, num_records)

customer_income = np.random.randint(20000, 120000, num_records)

customer_education = np.random.choice(education_levels, num_records)
```

```python
# Step 7: Generate book metadata

book_title_col = np.random.choice(book_titles, num_records)

author_col = np.random.choice(authors, num_records)

genre_col = np.random.choice(genres, num_records)
```

```python
# Step 8: Combine into dataset

df = pd.DataFrame({

    "Book_Title": book_title_col,
    "Author": author_col,
    "Genre": genre_col,

    "Price": prices,
    "Quantity_Sold": quantity_sold,
    "Revenue": revenue,
    "Cost": cost,
    "Profit": profit,
    "Profit_Margin (%)": profit_margin,

    "Customer_Age": customer_age,
    "Customer_Gender": customer_gender,
    "Customer_Region": customer_region,
    "Customer_Income": customer_income,
    "Customer_Education": customer_education
})

df.head()
```

```python
# Step 9: Inspect dataset

df.info()
df.describe()
```

```python
# Step 10: Save dataset

df.to_csv("LLM_BOOKS_Sales_Dataset.csv", index=False)

print("Dataset saved successfully.")
```

```python
total_revenue = df["Revenue"].sum()
total_profit = df["Profit"].sum()

print("Total Revenue:", total_revenue)
print("Total Profit:", total_profit)
```

```python
genre_profit = df.groupby("Genre")["Profit"].mean()

genre_profit.plot(kind="bar", title="Average Profit by Genre")
plt.show()
```

```python
region_sales = df.groupby("Customer_Region")["Revenue"].sum()

region_sales.plot(kind="bar", title="Revenue by Region")
plt.show()
```
